```python
#SSL certificate check
#using the function ssl.create_default_context(), and verify_mode set to bypass certificate
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
```

```python
# Reading the URL and passing it to BeautifulSoup
top100_url = 'https://www.gutenberg.org/browse/scores/top'
response = requests.get(top100_url)
response
```

```python
# Writing the function to check the web request status, if result is 200 then success all else is not
def web_request_status(response):
    if response.status_code == 200:
        print('Response is Successful')
    else:
        print('Response failed')
        
web_request_status(response)
```

```python
# Decoding the response and passing it to BeautifulSoup
cont = response.content.decode(response.encoding)
soup = BeautifulSoup(cont, 'html.parser')
```

```python
# Finding all the `href` tags and storing them in the list of links, then checking 
# what the list looks like
# Creating an empty list
href_list = []
for link in soup.find_all('a'):
    href_list.append(link.get('href'))
href_list[:30]
```

```python
# Using a regular expression to find the numeric digits in these links, then initializing the empty
# list to hold the file numbers
# first, creating the empty list
book_number = []
# creating a regular expression to find the numeric digits by using a for loop and then stripping the numbers
for i in range (19,119):
    link = href_list[i]
    link = link.strip()
    n = re.findall('[0-9]+', link)
    if len(n) == 1:
        book_number.append(int(n[0]))
        
print(book_number)
```

```python
# Using the `.text` object to find what the text object look like and printing the first 2k characters
print(soup.text[:2000])
```

```python
# Finding the top 100 ebooks
# Creating empty list to capture the 100 top ebooks
ebooks_list_temp=[]
# Creating a starting index to capture the top 100 ebooks using splitlines of soup.text
starting_index = soup.text.splitlines().index('Top 100 EBooks yesterday')
```

```python
# Looping 1-100 to add the strings in to the ebooks_list
for i in range(100):
    ebooks_list_temp.append(soup.text.splitlines()[starting_index+2+i])
ebooks_list_temp
```

```python
# Using regular expression to extract only text from the name strings and appending them to an empty list
ebooks_list =[]
for i in range(100):
    id1, id2=re.match('^[a-zA-z ]*', ebooks_list_temp[i]).span()
    ebooks_list.append(ebooks_list_temp[i] [id1:id2])
ebooks_list
```

```python
# Printing the list of titles one by one
for e in ebooks_list:
    print(e)
```

```python
# Openning JSON saved in folder and calling API key
with open ('APIkeys.json') as f:
    keys = json.load(f)
    omdbapi = keys['APIkey']
# Note, after successful run of the entire script will remove APIKeys file on desktop to not upload to github
```

```python
# Assigning the OMDB portal
omdb_url = 'http://www.omdbapi.com/?'
```

```python
# Creating a variable called `apikey` and assignment to the last portion of URL
apikey = '&apikey=' + omdbapi
```

```python
# Creating a utility function called `print_json` to print the movie data from the JSON file
def print_json(json_data):
    list_keys = ['Title', 'Year', 'Rated', 'Released', 'Runtime', 'Genre', 
                 'Director', 'Writer', 'Actors', 'Plot', 'Language', 'Country',
                 'Awards', 'Ratings', 'Metascore', 'imdbRating', 'imdbVotes', 'imdbID']
    for k in list_keys:
        if k in list(json_data.keys()):
            print(f'{k}: {json_data[k]}')
```

```python
# Writing a utility function to download a poster of the movie
def save_poster(json_data):
    title = json_data['Title']
    poster_url = json_data['Poster']
    poster_file_extension = poster_url.split('.')[-1]
    poster_data = urllib.request.urlopen(poster_url).read()
    savelocation = os.getcwd() + '\\' + 'Posters' + '\\'
    if not os.path.isdir(savelocation):
        os.mkdir(savelocation)
    filename = savelocation + str(title) + '.'+ poster_file_extension
    f1 = open(filename, 'wb')
    f1.write(poster_data)
    f1.close()
```

```python
# Search movie utility function
def search_movie(title):
    try:
        url = omdb_url + urllib.parse.urlencode({'t': str(title)}) + apikey
        print(url)
        uh = urllib.request.urlopen(url)
        data = uh.read()
        json_data = json.loads(data)
        
        if json_data['Response'] == 'True':
            print_json(json_data)
            if json_data['Poster'] != 'N/A':
                save_poster(json_data)
        else:
            print("Error: ", json_data['Error'])
    except urllib.error.URLError as e:
        print(f'Error: {e.reason}')
        
```

```python
search_movie('Titanic')
```

```python
search_movie('Random_error')
```

```python
# Will use the population data that I parsed from the web in Milstone 3 for graphs
url_ms3 = 'https://www.worldometers.info/world-population/population-by-country/'
header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}
```

```python
# Used the requests.get function to call the website (similar to APIs) and the table is now readable.
r = requests.get(url_ms3, headers=header)
dfs_web = pd.read_html(r.text)
```

```python
# defining the table to begin data wrangling process.
df_web_raw = dfs_web[0]
df_web_raw
```

```python
# Renaming columns for more clarity.
df_web = df_web_raw
df_web.columns = ['#', 'Country', 'Population', 'Yearly Change', 'Net Change', 'Density', 'Land Area', 
                  'Migrants', 'Fertility Rate', 'Median Age', 'Urban Population Percentage', 'World Share']
df_web
```

```python
# Replacing all NaN and N.A to 0
df_web = df_web.replace('N.A.', None)
df_web = df_web.fillna(0)
df_web
```

```python
# Line Graph
# Will use `.head(5)` to use top 5 lines in data from
df_web_line = df_web.head(5)
plt.plot(df_web_line['Country'], df_web_line['Population'])
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Line Chart: Country vs Population")
plt.show()
```

```python
# Bar Graph
df_web_bar = df_web.head(5)
plt.bar(df_web_bar['Country'], df_web_bar['Population'])
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Bar Chart: Country vs Population")
plt.show()
```

```python
# Histogram Graph
df_web_hist = df_web
plt.hist(df_web_hist['Population'], density=False, bins=100)
plt.ylabel('Probability')
plt.xlabel('Population')
plt.title("Population Count Histogram")
plt.show()
```

```python
# Density Plot
df_web_density = df_web
sns.distplot(df_web_density['Population'], bins=20, color = 'k')
```

```python
# Pie Plot
df_web_pie = df_web.head(10)
colors = sns.color_palette('pastel')[0:10]
plt.pie(df_web_pie['Population'], labels=df_web_pie['Country'], colors=colors, autopct='%.0f%%')
plt.show()
```
