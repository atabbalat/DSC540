```python
import pandas as pd
import numpy as np
import datetime
```

# Chapter 7

```python
# Reading the excel file for MetObjects data provided in the portal
df_raw = pd.read_csv('MetObjects.csv')
df_raw.head()
```

```python
# Will filter out missing data
# first will check for missing data even though it is visible from the table above
df_raw.isnull()
```

```python
# Now will use dropna() function to drop any row with missing data
df_drop_na = df_raw.dropna()
df_drop_na
```

```python
# Since dropping NA dropped every single rows, it means some of the columns are all NAs, by the look of
# dataframe above a column like 'River' may have everything being Na's, even though we can remove the column,
# a better approach to just replace it with 0
df_fill_na = df_raw.fillna(0)
df_fill_na
```

```python
# Will check df_fill_na for duplicated rows and removing them
df_fill_na.duplicated()
```

```python
# Removing duplicated rows by using drop_duplicates() function
df_duplicates = df_fill_na.drop_duplicates()
df_duplicates
# result shows there are no duplicates by checking the new total amount of rows.
```

# Chapter 8

```python
# will read 2 datasets of candy data 2015 and 2016 responses
df_candy_15 = pd.read_excel('CANDY-HIERARCHY-2015-SURVEY-Responses.xlsx')
df_candy_15.head()
```

```python
df_candy_16 = pd.read_excel('BOING-BOING-CANDY-HIERARCHY-2016-SURVEY-Responses.xlsx')
df_candy_16
```

```python
# Will combine both candy dataframes using pd.concat() function to see 2015 and 2016 in one table.
df_candy_conc = pd.concat([df_candy_15, df_candy_16])
df_candy_conc
```

```python
# Pivoting data to see what is in the 'Butterfinger' column
df_candy_conc_melt = df_candy_conc.stack()
df_candy_conc_melt
```

```python
# Indexing the dataframe to group eating apples side to side or like a freak followed by the listing 
# of the dress color that went viral
df_candy_ind = df_candy_conc.set_index(['Do you eat apples the correct way, East to West (side to side) or do you eat them like a freak of nature, South to North (bottom to top)?',
                                        '"That dress* that went viral a few years back - when I first saw it, it was ________"'])
df_candy_ind
```

# Chapter 10

```python
# We will go back and use MetObjects for this exercise and use df_duplicates
# We will group and see how many in each department got counted for
df_count = df_duplicates.groupby(['Department', 'Culture']).count()
df_count
```

```python
# Attempting to pivot by department where department is shown as columns and we are indexing culture
# Since we grouped by count all columns show the same count, and therefore we will use Object Number as
# the agg function
df_pivot = df_count.pivot_table(['Object Number'], index=['Culture'], columns='Department')
df_pivot = df_pivot.fillna(0)
df_pivot
```
