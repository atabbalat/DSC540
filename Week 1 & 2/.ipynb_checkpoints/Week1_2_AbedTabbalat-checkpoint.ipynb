###### Data Wrangling with Python: Activity 1 page 17

```python
# Creating a list of 100 random numbers.
r_list = [random.randint(0, 100) for x in range(0, 100)]
r_list
```

```python
# Creating a new list for numbers that are divisible by 3.
by3_list = [y for y in r_list if y % 3 == 0]
by3_list
```

```python
# Calculating the difference between the length of the two lists.
diff = len(r_list) - len(by3_list)
diff
```

```python
# Finding the difference variable 3 times.
diff_list = []
# will repeat all above in this exercise in the for loop.
loop_limit = 10

for i in range(0, loop_limit):
    r_list = [random.randint(0, 100) for x in range(0, 100)]
    by3_list = [y for y in r_list if y % 3 == 0]
    diff = len(r_list) - len(by3_list)
    diff_list.append(diff)

diff_list
```

```python
# Finding arithmetic mean
diff_avg = sum(diff_list) / len(diff_list)
diff_avg
```

```python
# Removing all new lines and symbols.
# Step 1 is to remove the breaks.
multiline_text = multiline_text.replace('\n', '')
multiline_text
```

```python
# Step 2, create a loop to remove all special characters.
clean_text = ''
for i in multiline_text:
    if i.isalnum():
        clean_text += i
    elif i == ' ':
        clean_text += i
    else:
        clean_text += ''
# Step 3, output.
clean_text
        
```

```python
# Converting string to list using split function.
clean_list = clean_text.split()
clean_list
```

```python
# Counting number of times the unique words.
len(clean_list)
```

```python
clean_dict = dict.fromkeys(clean_list)
clean_dict
```

```python
len(list(clean_dict.keys()))
```

```python
for w in clean_list:
    if clean_dict[w] is None:
        clean_dict[w] = 1
    else:
        clean_dict[w] += 1

clean_dict
```

```python
# Finding top 25 words from unique word counter.
clean_sorted = sorted(clean_dict.items(), key=lambda key_val_tuple:key_val_tuple[1], reverse=True)
clean_sorted[:25]
```

```python
for digits_3 in permutations(range(3)):
    print(digits_3)
    assert isinstance(digits_3, tuple)
```

```python
for digits_3 in permutations(range(3)):
    print(list(dropwhile(lambda x: x <= 0, digits_3)))
```

```python
for digits_3 in permutations(range(3)):
    stack = list(dropwhile(lambda x: x <= 0, digits_3))
    final = 0
    for i in range(0, len(stack)):
        final += (stack.pop() * (math.pow(10, i)))
        print(final)
```

###### Data Wrangling with Python: Activity 4 page 59

```python
def dict_csv (header, line):
    zipped = zip_longest(header, line, fillvalue=None)
    final_dict = {z[0]: z[1] for z in zipped}
    return final_dict
```

```python
with open ('sales_record.csv', 'r') as sr:
    init_line = sr.readline()
    header = init_line.replace('\n', '').split(',')
    for i, line in enumerate(sr):
        line = line.replace('\n', '').split(',')
        output = dict_csv(header, line)
        print(output)
        if i > 10:
            break
```
